## 5363. 친척 수
* https://coding-ai.aivle-edu.kr/Question/5363/View/1?fromPage=8

```python
# 1 (실패 2개)
def div(N): # N 의 약수들
    d = []
    for n in range(1, N + 1):
        if N % n == 0:
            d.append(n)
    return d

def gcd(A, B):
    while A != B and A > 0 and B > 0:
        if A >= B:
            A = A - B
        else:
            B = B - A
    
    return max(A, B)

N = int(input())
A = []
for _ in range(N):
    A.append(int(input()))

target = min(A)
A.remove(target)

temp = []
for x in A:
    temp.append(gcd(target, x))

D = div(min(temp))
print(D[0], end='')
for i in range(1, len(D)):
    print("", D[i], end='')
```

## 5362. 영화제
* https://coding-ai.aivle-edu.kr/Question/5362/View/1?fromPage=8
* https://www.acmicpc.net/problem/1671

```python
import random

def eval(A, B):
    win = 0
    for i in range(3):
        if A[i] >= B[i]:
            win += 1
        else:
            win -= 1
    
    if win == 3: # A가 B보다 전부 우세
        return 1
    elif win == -3: # B가 A보다 전부 우세
        return -1
    else: # 전부 우세는 아님
        return 0

def not0(A):
    check = 0
    for a in A:
        if a[3] != 0: # a[3] == 0 : 아직 한번도 심사 안됨
            check += 1
            
    if check == len(A):
        return True
    else:
        return False
    
def f5362(N, A):
    while not0(A) == False:
        randid = random.sample(list(range(N)), 2) # 0~N-1 중 랜덤 2개의 index 선택
        x, y = randid[0], randid[1]
        while A[x][3] < 0 or A[y][3] < 0: # 둘 중 하나가 이미 심사되서 떨어졌으면 재선택
            randid = random.sample(list(range(N)), 2)
            x, y = randid[0], randid[1]
            
        EV = eval(A[x], A[y])
        
        if EV == 1:
            A[x][3] += 1
            A[y][3] -= 1
        if EV == -1:
            A[x][3] -= 1
            A[y][3] += 1            

        count = 0
        for x in A:
            if x[3] > -1:
                count += 1
    
    return count


N = int(input())
A = []
for _ in range(N):
    t = [int(x) for x in input().split()]
    t.append(0) # 심사횟수
    A.append(t)
    
print(f5362(N, A))
```
